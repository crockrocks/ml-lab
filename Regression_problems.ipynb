# Linear Regression

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Load the diabetes dataset
from sklearn import datasets
diabetes_data = datasets.load_diabetes()
print(diabetes_data.DESCR)
```

```python
diabetes_df = pd.DataFrame(diabetes_data['data'] , columns = diabetes_data["feature_names"])
diabetes_df.head()
```

```python
diabetes_df["target"] = diabetes_data["target"]
diabetes_df.head()
```

```python
#creating X and y
X = diabetes_df.drop("target" , axis =1)
y = diabetes_df["target"]

np.random.seed(42)
# splitting the dataset into train and test dataset
from sklearn.model_selection import train_test_split
train_X , test_X , train_y , test_y = train_test_split(X , y , test_size = 0.2)
```

```python
# importing linear model and fitting the data
from sklearn.linear_model import LinearRegression
le_model = LinearRegression()
le_model.fit(train_X , train_y)
```

```python
#making prediction 
y_pred = le_model.predict(test_X)
y_pred
```

```python
result = pd.DataFrame({'Actual': test_y, 'Predict' : y_pred})
result
```

```python
#calculating the score
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error(test_y , y_pred)
```

```python
# r2 score
r2_score(test_y,y_pred)
```
